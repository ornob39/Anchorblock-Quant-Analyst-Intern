# Population & Sample

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
df['Population'] = [47, 48, 85, 20, 19, 13, 72, 16, 50, 60]
```

```python
df.head()
```

```python
df.plot.hist()
```

```python
sample_without_replacement = df["Population"].sample(n = 5, replace = False)
sample_without_replacement
```

```python
sample_with_replacement = df["Population"].sample(n = 5, replace = True)
sample_with_replacement
```

### Paramenters & Statistics

```python
population_mean = df["Population"].mean()
population_var = df["Population"].var(ddof= 0)
print(f"Poulation Mean: {population_mean}")
print(f"Population Variance: {population_var}")
```

```python
sample = df["Population"].sample(n = 10, replace = True)
sample_mean = sample.mean()
sample_var = sample.var(ddof = 1)
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_var}")
```

### Average of Uniform esimator

```python
sample_length = 500

sample_variance_collection = [df["Population"].sample(n = 10, replace = True).var(ddof = 1) for i in range(sample_length)]
```

```python
sample_variance_collection = pd.DataFrame(sample_variance_collection, columns = ["Sample Variance"])
```

```python
sample_variance_collection.plot.hist()
```
