# Variation of Sample

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
sample = pd.DataFrame(np.random.normal(10, 5, size = 30))

print(f"Sample mean is: {sample[0].mean()}")
print(f"Sample standard deviation is: {sample[0].std()}")
```

### Empirical Disrtibution of Mean

```python
meanlist = []

for _ in range(10000):
    sample = pd.DataFrame(np.random.normal(10, 5, size = 30))
    meanlist.append(sample[0].mean())
    
```

```python
collection = pd.DataFrame()
collection["meanlist"] = meanlist
```

```python
collection["meanlist"].hist(bins = 100, figsize=(15,8))
```

### Sampling from Arbitrary Distribution

```python
sample_size = 100
samplemeanlist = []
population = pd.DataFrame([1, 0, 1, 0, 1])

for _ in range(10000):
    sample = population.sample(sample_size, replace = True)
    samplemeanlist.append(sample[0].mean())
collection = pd.DataFrame()
collection["meanlist"] = samplemeanlist
collection.hist(bins = 100, figsize=(15,8))
    
```
