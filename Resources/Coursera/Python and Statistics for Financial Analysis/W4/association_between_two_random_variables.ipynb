```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

```python
df = pd.read_csv('../W1/data/housing.csv', index_col= 0)
df.head()
```

```python
df.cov()
```

```python
cols = df.columns.to_list()
cols
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale=1.5)
sns.heatmap(df[cols].cov(), annot=True, fmt='.2f', cmap='coolwarm')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(6,6))
sns.set(font_scale=1.5)
sns.heatmap(df[cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.tight_layout()
plt.show()
```
